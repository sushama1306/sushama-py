```python
def add_numbers(x,y):
    sum = x + y
    return sum
num1 = 10
num2 = 15
print("The sum is", add_numbers(num1,num2))
```

```python
def add_numbers(x,y):
    sum = x + y
    return sum
num1 = int(input("Enter first number : "))
num2 = int(input("Enyer second number : "))
print("The sum is : ", add_numbers(num1,num2))
```

```python
def find_square(num1):
    result = num1 * num1
    return result
num1 = int(input("Enter the Number : "))
print("The square of the num1 is :", find_square(num1))
```

```python
square = find_square(7)
print("Square of the no. is : ", square)
```
