```python
x=input("Enter number one : ")
y=input("Enter numer two : ")
try:
    z=int(x)/int(y)
except Exception as e:
  print("Expection occured :",e)
z=None
print("Division is :",z)
```

```python
x=input("Enter number one : ")
y=input("Enter numer two : ")
try:
    z=int(x)/int(y)
except ZeroDivisionError as e:
  print("Division by zero exception")
z=None
print("Division is :",z)
```

```python
x=input("Enter number one : ")
y=input("Enter numer two : ")
try:
    z=x/int(y)
except ZeroDivisionError as e:
  print("Division by zero exception")
z=None
print("Division is :",z)
```

```python
x=input("Enter number one : ")
y=input("Enter numer two : ")
try:
    z=x/int(y)
except ZeroDivisionError as e:
    print("Division by zero exception")
    z=None
except Exception as e:
    print("exception type :",type(e))
    z=None
    print("Division is :",z)
```

```python
x=input("Enter number one : ")
y=input("Enter numer two : ")
try:
    z=x/int(y)
except ZeroDivisionError as e:
    print("Division by zero exception")
    z=None
except TypeError as e:
    print("TypeError exception")
    z=None
    print("Division is :",z)
```
